```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass  # to get the password without showing the input
password = getpass.getpass()
```

Create a Python connection with SQL database and retrieve the results of the last two queries (also mentioned below) as dataframes:


```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
```

```python
#Check the number of rentals for each customer for May.
No_of_Rentals_may = '''              
              select customer_id, count(rental_id) as Rentals_M from rentals_may
              group by customer_id
              order by Rentals_M desc;
              '''
data_may = pd.read_sql_query(No_of_Rentals_may, engine)
data_may
```

```python
#Check the number of rentals for each customer for June
No_of_Rentals_june = '''              
              select customer_id, count(rental_id) as Rentals_J from rentals_june
              group by customer_id
              order by Rentals_J desc;
              '''
data_june = pd.read_sql_query(No_of_Rentals_june, engine)
data_june
```

```python
#Write a function that checks if customer borrowed more or less films in the month of June as compared to May.
mergedf = data_june.merge(data_may, how='left')
mergedf
```

```python
mergingdf = mergedf.dropna()
mergingdf
```

```python
#Function
mergingdf['Comparison'] = mergingdf.apply(lambda x : 'Rentals_June' if x['Rentals_J'] > x['Rentals_M'] else 'Rentals_May', axis=1)
mergingdf
```
